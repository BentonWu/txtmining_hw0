```python
#table是表格料來源
#tag是想要統計的欄位,如被保人性別

def Count(table,tag):
    data = table
    tag = str(tag)
    #取出所有業務來源
    business_source=data[tag].unique()
    business_source_list=business_source.tolist()
    #business_source_list.remove(business_source_list[0])
    business_source_count = pd.DataFrame(business_source_list)
    business_source_count=[]
    for i in range(len(business_source_list)):
        temp = data[data[tag].isin([business_source_list[i]])]
        temp_count = len(temp)
        a = [temp_count]
        business_source_count = business_source_count+a
        #rint("業務來源: "+str(business_source_list[i]) +" 的個數是" + str(temp_count))
    business_source_ratio=[]
    for i in range(len(business_source_count)):
        a = [business_source_count[i]/sum(business_source_count)]
        business_source_ratio = business_source_ratio+a
    business_source_dict = {
                "Types":business_source,
                "Counts": business_source_count,
                "Ratio" : business_source_ratio,
    }
    business_source_table = pd.DataFrame(business_source_dict)
    return(business_source_table)
```

```python
#切割日期的function
def cutter (start,end,data,column_name):
    down_side_mask = data[column_name]>=start
    up_side_mask = data[column_name]<=end
    masked_table = data[(up_side_mask&down_side_mask)]
    return(masked_table)
```
