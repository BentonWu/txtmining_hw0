```python
#貌似沒用到

def new_customer_numbers(start,end,data):
    #載入分割的表
    masked_table,before_table,in_table,before_end_table,before_start_table,before_end_table = cutter (start,end,data,'int格式起保日期')
    #比較新舊人的名單
    before_list=before_table['要保人ID'].unique()
    in_list = pd.Series(in_table['要保人ID'].unique())
    new_customer=in_table['要保人ID'].isin(before_list)
    #計算人數
    new_customer_count= len(in_table[new_customer])
    old_customer_count = len(in_table)-new_customer_count
    new_customer_ratio = new_customer_count/len(in_table)
    old_customer_ratio = old_customer_count/len(in_table)
    #print('新客戶數量是' +str(new_customer_count)+ '占比為' + str(new_customer_ratio))
    #print('舊客戶數量是' +str(old_customer_count)+ '占比為' + str(old_customer_ratio))
    return(new_customer_count,old_customer_count,new_customer_ratio,old_customer_ratio)
```

```python
#用日期把table切出
def cutter (start,end,data,column_name):
    day_time=Data_IF[column_name]
    #day_time_str_index = Data_IF[column_name].index
    #for i in range(len(day_time_str_index)):
        #把str格式的日期換成可以運算的格式
        #q=day_time[day_time_str_index[i]]
    before_mask = day_time<datetime.strptime(str(start),'%Y%m%d')
    down_side_mask = day_time>=datetime.strptime(str(start),'%Y%m%d')
    up_side_mask = day_time<=datetime.strptime(str(end),'%Y%m%d')
    
    masked_table = data[(up_side_mask&down_side_mask)]
    before_start_table = data[(before_mask)]
    before_end_table = data[(up_side_mask)]
    
    return(masked_table,before_start_table,before_end_table)
```

```python
def counting(start,end,data):

    
    
    #載入由cutter函數切出來的五個表
    #masked table是兩段期間切出
    #before start table是初始日之前的表
    #before end table 是截止日之前的表
    masked_table,before_start_table,before_end_table=cutter(start,end,data,'起保日')
    
    #用來計算年度有效總保人
    total_before_end_date_list=before_end_table.要保人ID.nunique()

    #計算有購單的要保人數
    total_bought_numbers=masked_table.要保人ID.nunique()

    #舊客戶名單
    old_customer_list = before_start_table.要保人ID.unique()
    #舊客戶人數
    old_customer_numbers=before_start_table.要保人ID.nunique()
    #舊客戶為True新客戶為False
    old_customer_logic = masked_table['要保人ID'].isin(old_customer_list)

    #建立時間內新舊客戶table
    old_customer_table=masked_table[(old_customer_logic)]
    new_customer_table=masked_table[(~old_customer_logic)]

    #處理舊客戶資料
    old_customer_list = old_customer_table.要保人ID.unique()
    old_policy_numbers = len(old_customer_table)

    #處理新客戶資料
    new_customer_list = new_customer_table.要保人ID.unique()
    new_policy_numbers = len(new_customer_table)
    
    #計算各種數值
    
    #1.有效總要保人數
    effective_ppl = total_before_end_date_list
    #2.期間有購單人數
    bought_intime_ppl = total_bought_numbers
    #3.期間購買保單數
    bought_intime_policy = len(masked_table)
    #4.新客戶人數
    new_ppl= len(new_customer_list)
    #5.新客戶購買保單數
    new_policy = new_policy_numbers
    #6.舊客戶人數
    old_ppl = len(old_customer_list)
    #7.舊客戶購買保單數
    old_policy = old_policy_numbers
    #8.舊保戶再購率 = 舊客戶再購人數/舊客戶人數
    return_ratio = old_ppl/old_customer_numbers


    #輸出
    print(str(end)+' 前有效總要保人數' +str((effective_ppl))+'人')
    print(str(start)+' 到 '+str(end)+' 之間有購單的要保人數'+str(bought_intime_ppl)+'人，總購單數'+str(bought_intime_policy)
    print('新客戶人數為:'+str(new_ppl)+'人，新購保單數為:'+str(new_policy))
    print('舊客戶再購人數為:'+str(old_ppl)+'人，再購保單數為:'+str(old_policy))
    print('既有客戶再購率為: '+str(len(old_customer_list)/len(total_before_end_date_list)))

    return(effective_ppl,bought_intime_ppl,bought_intime_policy,new_ppl,new_policy,old_ppl,old_policy,return_ratio)
```
